# Convert the required columns value into int datatypes

```python
# Class column having string value need to convert it int

df['Class'].replace(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica' ], [1,2,3], inplace=True)
```

```python
df.head(5)
```

# Select the Feature variale and Target var

```python
# Feature variable
X = df.drop(['Class'], axis = 1)
X.head(10)
```

```python
# Traget variable is Class
y = df['Class']
y.head(8)
```

```python
# now we can split the data into train and test dataset

from sklearn.model_selection import train_test_split
```

# Split the data 80% is train and 20% is Test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Now we can create the  RandomForestClassifier Module

```python
rfc = RandomForestClassifier(n_estimators= 1000, random_state=1)

accuracies = {}
```

```python
rfc.fit(X_train,y_train)
```

```python
# acc = rfc.score(X_test, y_test) *100
acc = rfc.score(X_test, y_test) 
acc
```

```python
accuracies['Random Forest'] = acc

print('random forest classifier accuracy: {:.2f}%'.format(acc))
```
